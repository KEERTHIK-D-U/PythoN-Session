```python
# ============================================
# SALES DATA ANALYTICS & VISUALIZATION
# ============================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a synthetic sales dataset
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
              'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Electronics': [120000, 135000, 142000, 128000, 150000, 155000,
                    160000, 170000, 168000, 180000, 190000, 200000],
    'Clothing': [80000, 95000, 100000, 85000, 97000, 105000,
                 110000, 115000, 120000, 130000, 140000, 145000],
    'Groceries': [60000, 65000, 70000, 72000, 74000, 76000,
                  80000, 83000, 85000, 88000, 91000, 95000]
}

df = pd.DataFrame(data)
print(" Initial Sales Data:\n", df.head(), "\n")

#  Compute total and average sales using NumPy
df['Total_Sales'] = np.sum(df[['Electronics', 'Clothing', 'Groceries']], axis=1)
df['Average_Sales'] = np.mean(df[['Electronics', 'Clothing', 'Groceries']], axis=1)

print(" Monthly Summary:\n", df[['Month', 'Total_Sales', 'Average_Sales']], "\n")

#  Identify month with highest total sales
best_month = df.loc[df['Total_Sales'].idxmax(), 'Month']
print(f" Highest total sales in: {best_month}\n")

# ============================================
# MATPLOTLIB VISUALIZATIONS
# ============================================

plt.style.use('seaborn-v0_8-darkgrid')  # nice background

#  Line Chart – Monthly Sales Trend
plt.figure(figsize=(8, 4))
plt.plot(df['Month'], df['Electronics'], marker='o', label='Electronics')
plt.plot(df['Month'], df['Clothing'], marker='o', label='Clothing')
plt.plot(df['Month'], df['Groceries'], marker='o', label='Groceries')
plt.title("Monthly Sales Trend by Category")
plt.xlabel("Month")
plt.ylabel("Sales (in ₹)")
plt.legend()
plt.tight_layout()
plt.show()

#  Bar Chart – Total Monthly Sales
plt.figure(figsize=(8, 4))
plt.bar(df['Month'], df['Total_Sales'], color='coral', edgecolor='black')
plt.title("Total Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales (₹)")
plt.tight_layout()
plt.show()

# Pie Chart – Category Contribution in Annual Sales
category_totals = [
    df['Electronics'].sum(),
    df['Clothing'].sum(),
    df['Groceries'].sum()
]
labels = ['Electronics', 'Clothing', 'Groceries']

plt.figure(figsize=(6, 6))
plt.pie(category_totals, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Annual Sales Share by Category")
plt.tight_layout()
plt.show()

#  Scatter Plot – Average vs Total Sales
plt.figure(figsize=(7, 4))
plt.scatter(df['Average_Sales'], df['Total_Sales'], color='teal')
plt.title("Average vs Total Monthly Sales")
plt.xlabel("Average Sales (₹)")
plt.ylabel("Total Sales (₹)")
for i in range(len(df)):
    plt.text(df['Average_Sales'][i]+2000, df['Total_Sales'][i],
             df['Month'][i], fontsize=8)
plt.tight_layout()
plt.show()

#  Highlighting Trends – Moving Average Line
df['Moving_Avg'] = df['Total_Sales'].rolling(window=3).mean()

plt.figure(figsize=(8, 4))
plt.plot(df['Month'], df['Total_Sales'], color='blue', marker='o', label='Total Sales')
plt.plot(df['Month'], df['Moving_Avg'], color='red', linestyle='--', label='3-Month Moving Avg')
plt.title("Sales Trend with Moving Average")
plt.xlabel("Month")
plt.ylabel("Sales (₹)")
plt.legend()
plt.tight_layout()
plt.show()

#  Display Key Insights
print(" KEY INSIGHTS:")
print(f"- Average total sales per month: ₹{df['Total_Sales'].mean():.2f}")
print(f"- Category with highest annual sales: {labels[np.argmax(category_totals)]}")
print(f"- Total annual sales: ₹{df['Total_Sales'].sum():,.0f}")
```
