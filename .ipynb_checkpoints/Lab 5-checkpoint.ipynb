```python
# Simple Sales Analysis using Pandas & Matplotlib

import pandas as pd
import matplotlib.pyplot as plt

# ------------------------------
# Step 1: Create a simple dataset
# ------------------------------
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [15000, 18000, 17000, 22000, 24000, 20000],
    'Profit': [4000, 5000, 4800, 6200, 7000, 6500],
    'Units_Sold': [120, 140, 130, 170, 180, 160]
}

df = pd.DataFrame(data)
print(" Sales Data:\n", df, "\n")

#  Basic statistics
print(" Average Sales:", df['Sales'].mean())
print(" Total Profit:", df['Profit'].sum())
print(" Month with Highest Sales:", df.loc[df['Sales'].idxmax(), 'Month'])
print()

# Visualizations
plt.figure(figsize=(10, 6))

# Bar Chart - Sales per Month
plt.subplot(2, 2, 1)
plt.bar(df['Month'], df['Sales'], color='skyblue')
plt.title('Sales per Month')
plt.xlabel('Month')
plt.ylabel('Sales')

# Line Chart - Profit Trend
plt.subplot(2, 2, 2)
plt.plot(df['Month'], df['Profit'], color='green', marker='o')
plt.title('Profit Trend')
plt.xlabel('Month')
plt.ylabel('Profit')

# Pie Chart - Units Sold Share
plt.subplot(2, 2, 3)
plt.pie(df['Units_Sold'], labels=df['Month'], autopct='%1.1f%%', startangle=90)
plt.title('Units Sold Share by Month')

# Combined Chart (Sales vs Profit)
plt.subplot(2, 2, 4)
plt.plot(df['Month'], df['Sales'], label='Sales', marker='o')
plt.plot(df['Month'], df['Profit'], label='Profit', marker='x')
plt.title('Sales vs Profit')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()

plt.tight_layout()
plt.show()
```
